## Import the Data

```python
# Directory Path
train_data_dir = 'chest_xray/train/'
val_data_dir = 'chest_xray/val/'
test_data_dir = 'chest_xray/test/'
```

```python
train = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
val = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
test = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

```python
# Get all the data in the directory chest_xray/train
train_generator = train.flow_from_directory(
                  train_data_dir, 
                  target_size = (384, 384),
                  batch_size=32, 
                  class_mode='binary')
```

## Building a Baseline Model

```python
base_model = keras.models.Sequential([
    keras.layers.InputLayer(input_shape=[384, 384,3]),
    keras.layers.Flatten(),
    keras.layers.Dense(300, activation='relu'),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
base_model.summary()
```

```python
# Score trained model.
scores = base_model.evaluate_generator(test_generator)
print('Test loss:', scores[0])
print('Test accuracy:', scores[1])
```

## Plotting a Confusion Matrix and an ROC curve

Accuracy is not always the best way to evaluate a classifier. Plotting a confusion matrix can be a more successful measure. To compute a confusion matrix, we first need to have a set of predictons so that they can be compared to the actual targets.

Just like cross_val_score() function, cross_val_predict() performs K-fold cross-validation, but instead of returning the evaluation scores, it reutrns the predictions made on each test fold. This means you get a clean prediction for each instance in the training set. ('clean' meaning that the prediction is made by a model that never saw the data during training).

Each row in a confusion matrix represents an actual class, while each column represents a predicted class. The first row of this matrix considers the negative class. The second row represents the positive class. The accuracy of positive predictions is Precision; True Positives / True Positives + False Positives. Recall, or sensitivity/true positive rate is the ratio of positive instances that are correctly dected by the classifier; True Positives / True Positives + False Negatives.

```python
confusion_matrix = confusion_matrix(pred_df['Has Pneumonia'], pred_df['Prediction'])
sns.heatmap(confusion_matrix, xticklabels=('Normal', 'Pneumonia'), annot=True, fmt='.3g')
```

```python
tpr, fpr, thresh = roc_curve(pred_df['Has Pneumonia'], pred_df['Prediction'])
roc = pd.DataFrame([tpr, fpr]).T
roc.plot(figsize=(10,8), color='darkorange', legend=None)
sns.set_style('darkgrid', {'axes.facecolor':'0.9'})
plt.figure(figsize=(10,8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve')
plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC) curve')
```

## A more complex Neural Network

```python
model = Sequential()
model.add(InputLayer(input_shape=[384, 384,3])),
model.add(Conv2D(32, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
def process_images(image_list):
    
    x = []
    y = []
    
    for image in image_list:
        x.append(cv2.resize(cv2.imread(image, cv2.IMREAD_COLOR), (nrows, ncolumns), interpolation=cv2.INTER_CUBIC))
        # Retrieve Labels
        if 'NORMAL' in image:
            y.append(0)
        elif 'PNEUMONIA' in image:
            y.append(1)
    return x, y
```

## Fine Tuning Neural Network Hyperparameters

```python
# Let's try combinations of hyperparameters and see which one works 
# best on the validation set (or using K-fold cross-val).
def build_model(n_hidden=1, n_neurons=30, learning_rate=3e-3, input_shape=[8]):
    model = keras.models.Sequential()
    model.add(keras.layers.InputLayer(input_shape=input_shape))
    for layer in range(n_hidden):
        model.add(keras.layers.Dense(n_neurons, activation='relu'))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.SGD(lr=learning_rate)
    model.compile(loss='mse', optimizer=optimizer)
    return model
```

```python
# Now let's build a KerasRegressor on this build_model function
keras_reg = keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

## Creating a Pipeline, GridSearch for optimal Clustering

```python
# Create a function that returns a model, taking as parameters things you
# want to verify using cross-valdiation and model selection
def create_model(optimizer='adam', kernel_initializer='glorot_uniform', dropout=0.2):
    model = Sequential()
    model.add(InputLayer(input_shape=[64,64,3]))
    model.add(Dense(64, activation='relu', kernel_initializer=kernel_initializer))
    model.add(Dropout(dropout))
    model.add(Dense(2, activation='sigmoid', kernel_initializer=kernel_initializer))
    
    model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
    
    return model
```

```python
# wrap the model using the function you created
clf = KerasRegressor(build_fn=create_model,verbose=0)
```

```python
# Create the Pipeline
pipeline = Pipeline([
                    ('kmeans', KMeans(n_clusters=0.95)),
                    ('clf', clf)
])
```

```python
# Create the Parameter Grid
param_grid = dict(kmeans__n_clusters=range(2, 5))
```

```python
# Perform the Grid Search Cross Validation
grid_clf = GridSearchCV(pipeline, param_grid, cv=3, verbose=2)

# Fit the Grid Search
grid_clf.fit(train_generator)
```

## Performance Measures

Measuring accuracy using cross-validation

```python
# The following does roughly the same thing as Scikit-Learn's cross_val_score()
# The StratifiedKFold class performs stratified sampling to produce folds
# that contain a representative ration of each class. At each iteration
# the code creates a clone of the classifier, it trains that clone on the 
# training folds, and makes predictions on the test fold. Then it counts
# the number of correct predictions and outputs the ration of correct predictions.


from skelarn.model_selection import StratifiedKFold
from sklearn.base import clone

skfolds = StratifiedKFold(n_splits=3, random_state=42)

for train_index, test_index in skfolds.split(xtrain, ytrain):
    clone_clf = clone(sgd_clf)
    xtrain_folds = xtrain[train_index]
    ytrain_folds = ytrain[train_index]
    xtest_fold = xtrain[test_index]
    ytest_fold = ytrain[test_index]
    
    clone_clf.fit(xtrain_folds, ytrain_folds)
    y_pred = clone_clf.predict(xtest_fold)
    n_correct = sum(y_pred == ytest_fold)
    print(n_correct / len(y_pred))
```

```python
# Let's use the cross_val_score to evaluated SGDClassifier using K-fold
# cross validation with three folds. Remember that K-fold cross-validation
# means splitting the training set into K folds, then making predictions
# and evaluating them on each fold using a model trained on the remaining folds

from sklearn.model_selection import cross_val_score

cross_val_score(sgd_clf, xtrain, ytrain, cv=3, sorting='accuracy')
```
